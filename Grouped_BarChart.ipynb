```
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np
```

```
N = 3
ind = np.arange(N)  # the x locations for the groups
width = 0.3       # the width of the bars

fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(224)

normal_images = [1261, 88, 234]
bar1 = ax.bar(ind, normal_images, width)

pneumonia_images = [3795, 88, 390]
bar2 = ax.bar(ind+width, pneumonia_images, width)

plt.ylim(0, 4500)
ax.set_ylabel('No. of Imgaes')
ax.set_xticks(ind+width/2.)
ax.set_xticklabels( ('Training Data', 'Validation Data', 'Test Data') )
ax.legend( (bar1, bar2), ('NORMAL', 'PNEUMONIA') )

#to display labels on bars
def autolabel(bars):
    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x()+bar.get_width()/2., 50+h, '%d'%int(h), ha='center', va='bottom')

autolabel(bar1)
autolabel(bar2)

plt.show()
```
